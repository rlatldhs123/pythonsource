```python
# score.txt 읽어서 평균과 합 출력
with open("./resource/score.txt", "r") as f:
    list1 = f.readlines()
    hap = 0
    for num in list1:
        # print(num, type(num), end="")
        hap += int(num)
print()        
print("점수 총계 : ", hap)
print("점수 평균 : ", hap / len(list1))
```

```python
score = []
with open("./resource/score.txt", "r") as f:
    for num in f:
        # print(num, end="")
        score.append(int(num))

print(sum(score))
print(sum(score) / len(score))
```

```python
# score.txt 읽어서 총합과 평균을 구한 후 result.txt 작성
# result.txt
# 총합 : 520
# 평균 : 86.67

score = []
with open("./resource/score.txt", "r") as f:
    for num in f:
        # print(num, end="")
        score.append(int(num))

with open("./resource/result.txt", "w", encoding="utf-8") as f1:
    f1.write("총합 : %d\n" % sum(score))
    f1.write("평균 : %.2f\n" % (sum(score) / len(score)))
```

```python
# info.txt
# 차다, 170, 48
import random

strings = list("가나다라마바사아자차카타파하")

with open("./resource/info.txt", "w", encoding="utf-8") as f:
    for i in range(500):
        name = random.choice(strings) + random.choice(strings)
        height = random.randrange(150,200)
        weight = random.randrange(50,100)
        f.writelines(f"{name}, {height}, {weight}\n")
```

```python
# info.txt 읽어서 BMI 지수 계산
# bmi = weight / (height / 100 ** 2)
# bmi >= 25 과체중, bmi >= 18.5 정상체중, 저체중

# 카다, 193, 72
# 이름 : 카다
# 키 : 193
# 몸무게 : 100
# BMI : 25
# 결과 : 정상체중

with open("./resource/info.txt", "r", encoding="utf-8") as f:
    for row in f:
        # print(row, end="")
        name, height, weight = row.strip().split(", ")
        print(name,height,weight)
        bmi = int(weight) / ((int(height) / 100) ** 2)
        result = None
        if 25 <= bmi:
            result = "과체중"
        elif 18.5 <= bmi:
            result = "정상체중"
        else:
            result = "저체중"
        # print(f"이름 : {name}\n")
        # print(f"키 : {height}\n")
        # print(f"몸무게 : {weight}\n")
        print("\n".join(["이름 : {}","키 : {}","몸무게 : {}","BMI : {}","결과 : {}"]).format(name,height,weight,bmi,result))
        print()
```

```python
content = None

while True:
    no = int(input("1. 암호화 | 2. 복호화 | 3. 종료 중 원하는 메뉴 번호 입력 : "))

    if no == 1:
        # origin.txt 읽은 후 content 변수에 담기
        # content 변수에 담은 값을 chr(ord(가) + 100) 변경 후 파일에 작성 : encryption.txt
        with open("./resource/origin.txt", "r", encoding="utf-8") as f:
            content = f.read()

        with open("./resource/encryption.txt", "w", encoding="utf-8") as f:
            for c in content:
                f.write(chr(ord(c) + 100))
    elif no == 2:
        # encryption.txt 읽어온 후 원래의 문자로 출력
        with open("./resource/encryption.txt", "r", encoding="utf-8") as f:
            content = f.read()
            for i in range(0, len(content)):
                print(chr(ord(content[i])-100), end="")
    else:
        break
```
