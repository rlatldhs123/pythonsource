### requests 라이브러리 사용하기

- HTTP 요청 시 사용
- urllib.request 더 직관적이고 간단한 방법 제공
- 디코딩도 적절히 해줌
- json 처리가 편함

```python
import requests
```

```python
response=requests.get("https://www.naver.com")
print("status_code======================================")
print(response.status_code)
print("encoding========================================")
print(response.encoding)
print("text=============================================")
print(response.text)
```

```python

# 세션 + with => close() 호출 안해도 됨






with requests.Session() as s:

    response=s.get("https://www.naver.com")

    print(response.text)

```


> 다양한 http 메소드들도 지원을 하고 있음


- get() 
- post() 
- put()
- delete()

```python

url="https://httpbin.org/get"


with requests.Session() as s:

    response=s.get(url)

    print(response.text)
```

```python
url="https://httpbin.org/post"

# 폼에 삽입되는 데이터  dict 구조로 생성

data ={"name": "hong"}


with requests.Session() as s:

    response=s.post(url,data=data)

    print(response.text)
```

```python
url="https://httpbin.org/put"

# 폼에 삽입되는 데이터  dict 구조로 생성

data ={"name": "hong"}


with requests.Session() as s:

    response=s.put(url,data=data)

    print(response.text)
```

```python
url="https://httpbin.org/delete"

# 폼에 삽입되는 데이터  dict 구조로 생성




with requests.Session() as s:

    response=s.delete(url)

    print(response.text)
```

```python
url="https://httpbin.org/get"


# 파라메터
# https://httpbin.org/get?name=hong

# payload ={"name":"hong","age":24}
payload ={"name":"hong","age":24,"key":["value1","value2"]}




with requests.Session() as s:

    response=s.get(url,params=payload) # 파라메터 보낼때는  params 에다가 넣어주면 된다

    print(response.url)
```

```python
url = "https://jsonplaceholder.typicode.com/todos/"

with requests.Session() as s:

    response=s.get(url) # 파라메터 보낼때는  params 에다가 넣어주면 된다

    print(response.json())
    print(response.text)
    print(response.content)
```

```python
url = "https://jsonplaceholder.typicode.com/users"

with requests.Session() as s:

    response=s.get(url) # 파라메터 보낼때는  params 에다가 넣어주면 된다

    print(response.url)
#    print(response.text)
    print(response.json)





    for row in response.json():
        for k,v in row.items():
            print(f"{k}: {v}")
        print()            
```

```python
from fake_useragent import UserAgent





url="https://httpbin.org/get"

userAgent=UserAgent()
headers={"user-agent":userAgent.chrome}


# 파라메터
# https://httpbin.org/get?name=hong

# payload ={"name":"hong","age":24}
payload ={"name":"hong","age":24,"key":["value1","value2"]}




with requests.Session() as s:

    response=s.get(url,params=payload,headers=headers) # 파라메터 보낼때는  params 에다가 넣어주면 된다

    print(response.url)
```
